```python
import gymnasium as gym
from gym import spaces
import numpy as np
```

```python
from gym.spaces import space
class MyEnv(gym.Env):
    def __init__(self):
        super(MyEnv, self).__init__()

        self.action_space = gym.spaces.Discrete(4)

        self.observation_space = gym.spaces.Dict({
            'player_hp': spaces.Box(low=0, high=100, shape=(1,), dtype=np.float32),
            'opponent_hp': spaces.Box(low=0, high=100, shape=(1,), dtype=np.float32),
            'player_type': spaces.Discrete(15),
            'opponent_type': spaces.Discrete(15),

        })

        self.state = None

    def reset(self):
      self.state = {
          'player_hp': 100,
          'opponent_hp': 100,}
```

Create a list of Creatures, their attacks, stats, health, etcetera

```python
class Creature():
  def __init__(self,name,hp, attack,defense,speed,moveset,typing):
    ''' (Creature, str, int, int, int, int, dict)
        name: name of creature
        hp: health points
        attack: attack power
        defense: defense power
        speed: speed
        moveset: dictionary of moveset {move name: {typing, power, accuracy}
        type: type of creature
    '''
    self.name = name
    self.hp = hp
    self.attack = attack
    self.defense = defense
    self.speed = speed
    self.moveset = moveset
    self.typing = typing

  def __str__(self):
      return f"{self.name} (Type: {self.typing}) - HP: {self.hp}, Atk: {self.attack}, Def: {self.defense}, Spd: {self.speed}"


class Charflame(Creature):
  def __init__(self):
    super().__init__("Charflame",
                     hp=85
                     ,attack=84,
                     defense=78,
                     speed=100,
                     typing="Fire",
                     moveset={
                         "Flamethrower": {"Typing": "Fire", "Power": 95, "Accuracy":100},
                         "Earthquake": {"Typing": "Ground", "Power": 100, "Accuracy":100},
                         "Fire Blast": {"Typing": "Fire", "Power": 120, "Accuracy":85},
                         "Skull Bash": {"Typing": "Normal", "Power": 100, "Accuracy":100},
                     })


class Dragonking(Creature):
  def __init__(self):
    super().__init__("Dragonking",
                     hp=91,
                     attack=105,
                     defense=95,
                     speed=80,
                     typing="Dragon",
                     moveset={
                         "Slam": {"Typing": "Normal", "Power": 80, "Accuracy":75},
                         "Blizzard": {"Typing": "Ice", "Power": 120, "Accuracy":90},
                         "Dragon Punch": {"Typing": "Dragon", "Power": 90, "Accuracy":95},
                         "Fire Blast": {"Typing": "Fire", "Power": 120, "Accuracy":85},
                     })
```
